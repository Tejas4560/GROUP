#### 2. Create a histogram for each feature in the dataset to illustrate the feature distributions.

#### 3. Create a boxplot for each feature in the dataset.

```python
 plt.figure(figsize=(10, 6))
 sns.boxplot(df)
 plt.title('Boxplots of Features in Iris Dataset')
 plt.show()
```

#### 4. Outliers

```python
outliers = {}
for feature in df.columns:
    if df[feature].dtype in ['int64', 'float64']:  # Check if the column contains numeric data
        q1 = df[feature].quantile(0.25)
        q3 = df[feature].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        outliers[feature] = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]

print("Outliers:")
for feature, data in outliers.items():
    print(f"{feature}: {data.shape[0]} outliers")
```
