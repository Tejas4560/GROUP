-  Use the inbuilt dataset 'titanic'. The dataset contains 891 rows and contains information about the
passengers who boarded the unfortunate Titanic ship. Use the Seaborn library to see if we can fin 
any patterns in the dat
- .
 Write a code to check how the price of the ticket (column name: 'fare') for each passengeris 
distributed by plotting a histogram.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
## Histogram of Ticket Prices
plt.figure(figsize=(8,4))
sns.histplot(df['fare'],bins=30,kde=True)
plt.title("Distribution of Ticket Prices on Titanic")
plt.xlabel('Fare')
plt.ylabel('Frequency')
```

```python
## Distplot with rugplot for Age
plt.figure(figsize=(10,6))
sns.distplot(df['age'],rug=True)
plt.title("Distribution of Age with Rugplot")
plt.xlabel('Age')
plt.ylabel('Density')
```

```python
## Joinplot with scatter and kernal density estimation for fare and age
sns.jointplot(x='fare',y='age',data=df,kind='kde')
plt.title('Jointplot of Fare and Age')
```

```python
## Rugplot for age
sns.rugplot(df['age'])
plt.title('Rugplot for Age')
plt.xlabel('Age')
```

```python
# Violin Plot for fare and class
sns.violinplot(x='class',y='fare',data=df)
plt.title('Violin Plot of Fare by Class')
plt.xlabel('Class')
plt.ylabel('Fare')
```

```python
## Plot for sex and survival
sns.barplot(x='sex',y='survived',data=df)
plt.title('Bar plot of Survival by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Probability')
```

```python
## Box plot for Class and Age
sns.boxplot(x='class',y='age',data=df)
plt.title('Box Plot of Age and Class')
plt.xlabel('Class')
plt.ylabel('Age')
```
